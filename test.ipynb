Reverse list of k

```python
n = int(input("Enter the size of the list: "))
nums = []

for i in range(n):
    val = int(input(f"Enter element {i+1}: "))
    nums.append(val)

k = int(input("Enter the group size (K): "))

result = []


for i in range(0, len(nums), k):
   
    group = nums[i : i+k]
   
    result.extend(group[::-1])

print("Reversed in groups of K:", result)
 
```
