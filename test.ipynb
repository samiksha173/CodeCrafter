Sum of Digit

```python
n = int(input("Enter a Number :"))

digit_sum = 0 

while n> 0:
    digit = n%10

    digit_sum += digit

    n = n//10

print(digit_sum)
```

Count vowels

```python
str = input("Ente a string:")

list = ['a','i','e','o','u','A','I','E','O','U']
count = 0
for i in str :
    for j in list:
        if i == j : 
             count = count +1
print("Number of vowels:", count)
```
